```python
from nltk.corpus import indian
from nltk.tag import tnt
import nltk
train_data = indian.tagged_sents('hindi.pos')
tnt_pos = tnt.TnT()
tnt_pos.train(train_data)
```

```python
f = open("premchand.txt")
text = f.read()
f.close()
text = text.replace('।','')
text = text.replace('.','')
text = text.replace(',','')
text = text.replace(':','')
text = text.replace(';','')
text = text.replace('?','')
text = text.replace('!','')
text = text.replace('-','')
text = text.replace("'","")
text = text.replace('"','')

tok = nltk.word_tokenize(text)
```

```python
tagged = tnt_pos.tag(tok)
```
